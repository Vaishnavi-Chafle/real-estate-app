## 1. Society

```python
df["society"].value_counts()
```

```python
df["society"].value_counts().shape
```

```python
import re                                  
df["society"] = df["society"].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()  

# re is Python’s Regular Expressions module, used for pattern matching and string cleaning.
# str(name)  --> Sometimes a cell might be NaN or another type → converting avoids errors.
```

```python
df["society"].value_counts().shape
```

```python
df["society"] = df["society"].str.replace("nan","independent")
```

```python
df.head(5)
```

## 2.price

```python
df["price"].value_counts()
```

```python
df = df[df["price"] != "Price on Request"]
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if  x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
        
        
```

```python
df["price"] = df["price"].str.split(' ').apply(treat_price)
```

## 3.price_per_sqft

```python
df["price_per_sqft"].value_counts()
```

```python
df["price_per_sqft"] = df["price_per_sqft"].str.split("/").str.get(0).str.replace("₹",'').str.replace(',','').astype('float')
```
