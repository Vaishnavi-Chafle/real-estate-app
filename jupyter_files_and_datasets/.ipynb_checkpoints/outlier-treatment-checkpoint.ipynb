```python

# outliers on the basis of price column
sns.histplot(df['price'],kde=True)
```

```python
sns.boxplot(x=df["price"])
```

```python
# Identify the outliers

Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["price"] < lower_bound) | (df["price"] > upper_bound)]

outliers.sort_values('price',ascending=False).head(20)
```

```python
outliers.shape
```

```python
outliers["price"].describe()
```

```python
outliers.to_csv("outliers_price.csv")
# after scrolling this outliers in excel genuinly understood that , on the bases of price column that their are genuine outliers but their are somre
# data errors as well
```

## Price_per_sqft

```python
sns.kdeplot(df["price_per_sqft"],fill = True)
```

```python
sns.boxplot(x=df["price_per_sqft"])
```

```python
# identifying the outliers
Q1 = df["price_per_sqft"].quantile(0.25)
Q3 = df["price_per_sqft"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_sqft = df[(df["price_per_sqft"] < lower_bound) | (df["price_per_sqft"] > upper_bound)]
outliers_sqft.sort_values("price_per_sqft", ascending = False).head()
```

```python
outliers_sqft.shape
```

```python
outliers_sqft["price_per_sqft"].describe()
```

```python
outliers_sqft.to_csv("outliers_sqft.csv")

# 1) after scrolling this outliers in excel , it realize that the since the prices is in crodes the area has to be in the sqft but they are in yards
# 2) area - >"When I scanned down, I noticed that wherever the value is below 1000, it is in square yards, and wherever it is above 1000, 
# it is in square feet. 
```

```python
outliers_sqft
```

```python
outliers_sqft["area"] = outliers_sqft["area"].apply( lambda x : x * 9 if x < 1000 else x)
```

```python
outliers_sqft["price_per_sqft"] = round(outliers_sqft["price"] * 10000000/outliers_sqft["area"])
```
